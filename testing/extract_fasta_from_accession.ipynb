```python
%cd /home/neunna/workspace/nextclade/nextclade_a16/testing
from Bio import SeqIO
import numpy as np
import pandas as pd
from collections import defaultdict

seq_data =SeqIO.to_dict(SeqIO.parse("../results/sequences_with_ancestral.fasta", "fasta"))
accessions = pd.read_csv("A16_XuQ.csv", header=0)['accession'].tolist()
```

```python
## All names in accessions should be extracted in the seq_data and written to a new fasta file
with open("A16_XuQ_extracted.fasta", "w") as output_handle:
    for accession in accessions:
        if accession in seq_data:
            SeqIO.write(seq_data[accession], output_handle, "fasta")
        else:
            print(f"Warning: Accession {accession} not found in sequence data.")
```
